## **3) Analisi della distribuzione delle variabili**

## **4) Analisi della distribuzione dei dati**

Osservando i dati presentati sopra, si può vedere che le variabili `price`, `minimum_nights`e `calculated_host_listing_count`, si può verificare che c'è una distribuzione distorta che rende difficile visualizzare e comprendere i dati.

Per questa analisi sono state eseguite le seguenti procedure:

* Riepilogo statico che utilizza il metodo `describe()`
* Analizza i boxplots delle variabili selezionate.

```
# riassunto statistico delle variabili numeriche
df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].describe()
```

Osservando le informazioni fornite sopra, è possibile ottenere importanti informazioni sulla distribuzione e sulla presenza di alcuni valori indesiderati per l'analisi:

* Il valore del `price` prima del 75% dei dati è inferiore a 120 euro e il valore massimo mostrato è di 10.571,00 euro;
* Il valore massimo di `minimun_nights` è 1000, molto più alto del numero di giorni in un anno (365 giorni);
* Il prezzo minimo è zero, che può essere un errore, poi sappiamo che su Airbnb non c'è la possibilità di affittare un alloggio gratuito;

### Boxplot para Price:

```
# price
df.price.plot(kind='box', vert=False, figsize=(15, 3),)
plt.show()

# quantità di valore superiore a 300 euro
print("\nprice: valori superiori a 300 euro")
print("{} entrate".format(len(df[df.price > 300])))
print("{:.4f}%".format((len(df[df.price > 300]) / df.shape[0])*100))

# df.price.plot(kind='box', vert=False, xlim=(0,1000), figsize=(15,3));
```

### Boxplot para minimum_nights

```
# minimum_nights
df.minimum_nights.plot(kind='box', vert=False, figsize=(15, 3))
plt.show()

# quantità di valore superiore a 30 giorni para minimum_nights
print("minimum_nights: valori superiori a  30:")
print("{} entrate".format(len(df[df.minimum_nights > 30])))
print("{:.4f}%".format((len(df[df.minimum_nights > 30]) / df.shape[0])*100))
```

### Trattamento dei dati

Dopo aver controllato e verificato la presenza di *outliers*, rimuoveremo queste informazioni incoerenti in `price` e `minimum_nights`.

Rimuoveremo anche i valori zero per `price`.

Stamperemo nuovamente i grafici di distribuzione.




```
# rimuovere outliers in un nuovo dataframe
df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price > 1000].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 30].index, axis=0, inplace=True)

#rimuovendo valori di price uguali a zero
df_clean.drop(df_clean[df_clean.price == 0].index, axis=0, inplace=True)

# stampa l'istogramma delle variabili numeriche
df_clean.hist(bins=15, figsize=(15,10));
```

## **5) Prezzi medi di affitto**


In questa fase vengono presentati i valori medi in relazione agli alloggi disponibili:


```
# vedere la media della colonna `price`
print("Media in euro: " + str(df_clean['price'].mean()))
```

Il prezzo medio di locazione secondo i dati estratti è di 88,34 euro.

## **6) Analisi dei tipi di proprietà più affittati su Airbnb**

La variabile `room_type` indica il tipo di proprietà pubblicizzata. In questo modo è possibile identificare il numero di tipologie di immobili offerti e le loro quantità.

```
# quantità di ogni tipo di proprietà disponibile
df_clean.room_type.value_counts()
```

```
# percentuale di ogni tipo di proprietà disponibile
df_clean.room_type.value_counts() / df_clean.shape[0]
```

## **7) Analisi delle località più costose**

Raggruppando le informazioni di `neighbourhood_group` e `price` è possibile identificare la disposizione dei prezzi di affitto per quartieri a Roma.




```
df_clean.groupby(['neighbourhood']).price.mean().sort_values(ascending=False)[:10]
```

In questo modo è possibile verificare che i quartieri più cari si trovano nelle regioni del Centro Storico vicino ai punti principali della città.

<center><img alt="Analisando Airbnb" width="40%" src="https://upload.wikimedia.org/wikipedia/commons/thumb/0/0b/Mappa_dei_quartieri_di_Roma.svg/1200px-Mappa_dei_quartieri_di_Roma.svg.png"></center>



Poiché le latitudini e le longitudini sono fornite per le proprietà, è possibile tracciare ogni punto. Per questo vengono considerati `x = longitudine` e `y = latitudine`.

```
# stampa le proprietà per latitudine-longitudine

fig, ax = plt.subplots()

df_clean.plot(kind="scatter", x='longitude', y='latitude', alpha=1, c=df_clean['price'], s=8,
              cmap=plt.get_cmap('jet'), figsize=(15,10), ax=ax)
ax.set_title('Mappa termica di Roma: Latitudine x Longitudine x Prezzo',fontsize= 18)

plt.show()


```

## **8) Ci sono opzioni più economiche in posizioni privilegiate?**

Abbiamo potuto verificare che i prezzi più cari si trovano vicino ai punti turistici più rilevanti della città, ma ci sono buone opportunità per soggiornare in queste regioni.

Per analizzare questo, controlleremo la distribuzione dei valori per regione e cercheremo di identificare prezzi di alloggio interessanti che possono essere considerati *outlier* in relazione alla distribuzione dei prezzi della regione.

Quartieri selezionati:

* I Centro Storico

```
 sel_csto = df_clean.loc[df_clean['neighbourhood'] == 'I Centro Storico']
```
